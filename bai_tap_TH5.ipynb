### Câu 1b


```python
import numpy as np

A = np.array([[0,1,1], [0,-2,1],[2,1,1],[2,-2,1],[4,1,1],[6,-2,1]])
B = np.array([4.3,1.3,-5.9,-9.4,-15.8,-18.9])

temp = np.linalg.det(A.T @ A)
if temp == 0:
    print("A^T A is singular")
    exit(1)
else:
    print("det(A.T@A) = {}".format(temp))

X_star = np.linalg.inv(A.T @ A) @ A.T @ B
a, b, c = X_star
print(f"a: {a}")
print(f"b: {b}")
print(f"c: {c}")

r = np.linalg.norm(A @ X_star - B)
print(f"r(D): {r}")

a = round(a,4)
b = round(b,4)
c = round(c,4)
print(f"a: {a}")
print(f"b: {b}")
print(f"c: {c}")
X = np.array([a,b,c])
r = np.linalg.norm(A @ X - B)
print(f"r(D): {r}")
```

### Câu 2


```python
import numpy as np

x1 = np.array([2, 2, 3.06, 3.06, 4, 2.06, 2.06])
x2 = np.array([3, 6.07, 3.55, 6.97, 6, 3.62, 6.16])
b = np.array([3, 3, 3.89, 10.25, 13, 3.94, 3.53])

x1_x2 = x1 * x2
print(x1_x2)

A = np.vstack((np.ones(x1.shape[0]), x1, x2, x1_x2)).T

temp = np.linalg.det(A.T @ A)
if temp == 0:
    print("A^T A is singular")
    exit(1)
else:
    print("det(A.T@A) = {}".format(temp))

X_star = np.linalg.inv(A.T @ A) @ A.T @ b

beta_0, beta_1, beta_2, beta_3 = X_star

print(f"β0: {beta_0}")
print(f"β1: {beta_1}")
print(f"β2: {beta_2}")
print(f"β3: {beta_3}")

r = np.linalg.norm(A @ X_star - b)
print(f"r(D): {r}")

beta_0 = round(beta_0, 4)   
beta_1 = round(beta_1, 4)
beta_2 = round(beta_2, 4)
beta_3 = round(beta_3, 4)
print(f"β0: {beta_0}")
print(f"β1: {beta_1}")
print(f"β2: {beta_2}")
print(f"β3: {beta_3}")
X = np.array([beta_0, beta_1, beta_2, beta_3])
r = np.linalg.norm(A @ X - b)
print(f"r(D): {round(r, 4)}")
```

### Câu 3



$ln(y) = ax^2 + bx + c$


```python
import numpy as np


A = np.array([[0.25,0.5,1], [1,1,1], [2.25,1.5,1], [4,2,1]])
B = np.array([np.log(1.5), np.log(78.9),
             np.log(1751.1), np.log(59816.3)])
temp = np.linalg.det(A.T @ A)
if temp == 0:
    print("A^T A is singular")
    exit(1)
else:
    print("det(A.T@A) = {}".format(temp))

X_star = np.linalg.inv(A.T @ A) @ A.T @ B

a, b, c = X_star

print(f"a: {a}")
print(f"b: {b}")
print(f"c: {c}")


def model(x, a, b, c):
    return np.exp(a * (x**2) + b*x + c)


def calculate_r(A, Y, a, b, c):
    r = 0
    for i in range(len(A)):
        r += (Y[i] - model(A[i][1], a, b, c))**2
    return r**0.5


Y = np.array([1.5, 78.9, 1751.1, 59816.3])
r = calculate_r(A, Y, a, b, c)
print(f"r(D): {r}")

a = round(a, 4)
b = round(b, 4)
c = round(c, 4)
print(f"a: {a}")
print(f"b: {b}")
print(f"c: {c}")
r = calculate_r(A, Y, a, b, c)
print(f"r(D): {r}")
```

### Câu 4


$ln(y) = ln(a) + bx$

$$
\boldsymbol{A} =
\begin{bmatrix}
  0.4 && 1 \\ 
  0.8 && 1 \\ 
  1.2 && 1 \\ 
  1.6 && 1 \\
  2.0 && 1 \\
  2.3 && 1 
\end{bmatrix},

\boldsymbol{b} =
\begin{bmatrix}
  ln(750) \\
  ln(1000) \\
  ln(1400) \\
  ln(2000) \\
  ln(2700) \\ 
  ln(3750) 
\end{bmatrix}.$$

```python
import numpy as np


A = np.array([[0.4, 1], [0.8, 1], [1.2, 1], [1.6, 1], [2.0, 1], [2.3, 1]])
B = np.array([np.log(750), np.log(1000),
             np.log(1400), np.log(2000), np.log(2700), np.log(3750)])

temp = np.linalg.det(A.T @ A)
if temp == 0:
    print("A^T A is singular")
    exit(1)
else:
    print("det(A.T@A) = {}".format(temp))
    
X_star = np.linalg.inv(A.T @ A) @ A.T @ B

b, ln_a = X_star

print(f"ln_a: {ln_a}")
print(f"-> a:" , np.exp(ln_a))
print(f"b: {b}")


def model(x, a, b):
    return a* np.exp(b*x )


def calculate_r(A, Y, a, b):
    r = 0
    for i in range(len(A)):
        r += (Y[i] - model(A[i][0], a, b))**2
    return r**0.5


Y = np.array([750, 1000, 1400, 2000, 2700, 3750])
a = np.exp(ln_a)
r = calculate_r(A, Y, a, b)
print(f"r(D): {r}")

ln_a = round(ln_a, 4)
a = round(np.exp(ln_a), 4)
b = round(b, 4)
print(f"ln_a: {ln_a}")
print(f"a: {a}")
print(f"b: {b}")
r = calculate_r(A, Y, a, b)
print(f"r(D): {round(r, 4)}")
```

### Câu 5


$\dfrac{1}{y} = \dfrac{b+x}{ax} = \dfrac{b}{a}.\dfrac{1}{x} + \dfrac{1}{a}$

Let $Y = \dfrac{1}{y}, X = \dfrac{1}{x}, \alpha = \dfrac{b}{a}, \beta = \dfrac{1}{a}$ <br>
The equation will be: <br>
$Y = \alpha X + \beta$

Data table:
| X | 4/3 | 1/2   | 2/5 | 1/4 | 1/6   | 1/8 | 2/17  |
|---|-----|-------|-----|-----|-------|-----|-------|
| Y | 5/4 | 10/13 | 5/6 | 5/8 | 10/17 | 5/9 | 10/17 |

```python
import numpy as np

A = np.array([[4/3, 1], [1/2, 1], [2/5, 1], [1/4, 1], [1/6, 1], [1/8, 1], [2/17,1]])
B = np.array([5/4, 10/13, 5/6, 5/8, 10/17, 5/9, 10/17])

temp = np.linalg.det(A.T @ A)
if temp == 0:
    print("A^T A is singular")
    exit(1)
else:
    print("det(A.T@A) = {}".format(temp))
    
X_star = np.linalg.inv(A.T @ A) @ A.T @ B

alpha, beta = X_star

print(f"alpha: {alpha}")
print(f"beta: {beta}")

a = 1/beta
b = a * alpha
print(f"a: {a}")
print(f"b: {b}")


def model(x, a, b):
    return (a*x)/(b+x)


def calculate_r(X, Y, a, b):
    r = 0
    for i in range(len(A)):
        r += (Y[i] - model(X[i], a, b))**2
    return r**0.5


X = np.array([0.75, 2., 2.5, 4, 6, 8, 8.5])
Y = np.array([0.8, 1.3, 1.2, 1.6, 1.7, 1.8, 1.7])
r = calculate_r(X, Y, a, b)
print(f"r(D): {r}")

alpha = round(alpha, 4)
beta = round(beta, 4)
a = round(1/beta, 4)
b = round(a * alpha, 4)
print(f"alpha: {alpha}")
print(f"beta: {beta}")
print(f"a: {a}")
print(f"b: {b}")
r = calculate_r(X, Y, a, b)
print(f"r(D): {round(r, 4)}")
```

```python
import numpy as np


A = np.array([[1, 1], [5, 1], [6, 1], [9, 1], [2, 1], [3, 1], [3, 1], [
             10, 1], [7, 1], [5, 1], [5, 1], [5, 1], [8, 1], [7, 1]])
B = np.array([31, 12.5, 15.5, 6.7, 30, 21, 18.5,
             8.6, 9.0, 18.0, 11.0, 13.0, 11.0, 9.0])
B = np.log(B)

temp = np.linalg.det(A.T @ A)
if temp == 0:
    print("A^T A is singular")
    exit(1)
else:
    print("det(A.T@A) = {}".format(temp))

X_star = np.linalg.inv(A.T @ A) @ A.T @ B

b, ln_a = X_star

print(f"ln_a: {ln_a}")
print(f"-> a:", np.exp(ln_a))
print(f"b: {b}")

def model(x,a,b):
    return a * np.exp(b * x)

def calculate_r(A, B, a, b):
    r = 0
    for i in range(len(A)):
        r += (B[i] - model(A[i][0], a, b))**2
    return r**0.5


Y = np.array([31, 12.5, 15.5, 6.7, 30, 21, 18.5,
             8.6, 9.0,18.0, 11.0, 13.0, 11.0, 9.0])
r = calculate_r(A, Y, np.exp(ln_a), b)
print(f"r(D): {r}")
```
